```python
import numpy as np
```

```python
class LinearRegression:
    
    def __init__(self, fit_intercept=True):
        # fit_intercept(截距) : bool
        self.omega = None
        self.fit_intercept = fit_intercept
    
    def fit(self, X, y):
        if self.fit_intercept:
            X = np.c_[np.ones(X.shape[0]), X]
        pseudo_inverse = np.dot(np.linalg.inv(np.dot(X.T, X)), X.T)
        self.omega = np.dot(pseudo_inverse, y)
    
    def predict(self, X):
        if self.fit_intercept:
            X = np.c_[np.ones(X.shape[0], X)]
        return np.dot(X, self.omega)
```

```python
class RidgeRegression:
    
    def __init__(self, alpha=1, fit_intercept=True):
        # alpha : float
        self.omega = None
        self.alpha = alpha
        self.fit_intercept = fit_intercept
        
    def fit(self, X, y):
        if self.fit_intercept:
            X = np.c_[np.ones(X.shape[0]), X]
        # 和LInearRegression的不同
        A = self.alpha * np.eye(X.shape[1])
        pseudo_inverse = np.dot(np.linalg.inv(np.dot(X.T, X) + A),X.T)
        self.omega = np.dot(pseudo_inverse, y)
        
    def predict(self, X):
        if self.fit_intercept:
            X = np.c_[np.ones(X.shape[0]), X]
        return np.dot(X, self.omega)
```
